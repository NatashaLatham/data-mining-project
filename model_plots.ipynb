```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
feature_importances = pd.read_parquet('results/feature_importances.parquet')

ax = sns.barplot(data=feature_importances, x='importance', y='feature')
os.makedirs('results', exist_ok=True)
plt.title('Feature importances')
# for some reason, we need tight_layout, else the y-axis description is cut off...
plt.tight_layout()
# plt.savefig('results/feature_importances.pdf')
plt.show()
```

```python
feature_importances = pd.read_parquet('results/feature_importances.parquet')

feature = feature_importances['feature']
importance = feature_importances['importance']
std = feature_importances['std']

# Create the figure and axis objects
_, ax = plt.subplots()

# Create the horizontal barplot with errorbars
ax.barh(feature, importance, xerr=std, align='center')
plt.xlabel('feature_importance_')
plt.ylabel('feature')
plt.title('Feature importances')

# for some reason, we need tight_layout, else the y-axis description is cut off...
plt.tight_layout()
plt.savefig('results/feature_importances.pdf')
# Show the plot
plt.show()
```
